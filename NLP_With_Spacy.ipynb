```python
# Introduction to NLP and spaCy
import spacy
nlp = spacy.load('en_core_web_sm') # Model converts text into doc to store processed text
doc = nlp("Apple is looking at buying U.K. startup for $1 billion") # doc object
print([token.text for token in doc]) # Square brackets `[]` instantly build and store the entire list in memory (eager evaluation),
#while omitting them creates a memory-saving generator that processes items one at a time (lazy evaluation).

```
